```python
col_names = ["Episode Name", "Episode Text"]
df = pd.DataFrame(columns = col_names)
df
```

```python
for ep in episodes:
  ep_title = ep[:-4]
  file = open("SpongeBob_SquarePants_Transcripts/" + ep)
  ep_text = file.read()
  ep_text.close()
  ep_dict = {}
  ep_dict["Episode Name"] = ep_title
  ep_dict["Episode Text"] = ep_text
  #df.append(ep_dict, ignore_index = True)
  curr_df = pd.DataFrame(ep_dict, index = [0])
  df = pd.concat([df, curr_df], ignore_index = True, axis = 0)

df
```

```python
ratings = pd.read_csv('/content/drive/MyDrive/w266 Final Project Resources/binary_ratings.csv')
ratings["Episode Name"]  = ratings.Episode.str.replace(' ','')
ratings.head()
```

```python
df = df.merge(ratings, on = "Episode Name").drop(columns = "Unnamed: 0")
df.head()
```

```python


#df["Episode Text"] = df["Episode Text"].apply(lambda x: x + )
df['Rating']= '<'+df['Rating'].astype(str)+'>'
episode_texts = []
for i in range(len(df)):
  row = df.iloc[i]
  text = row["Episode Text"]
  new_text =  str(row["Rating"]) + " " + text
  episode_texts.append(new_text)

df["Episode Text"] = episode_texts

df["Episode Text"] = df["Episode Text"].apply(lambda x: '<Start Ep> ' + x)

df["Episode Text"] = df["Episode Text"].apply(lambda x: x + ' <End Ep>')

df

```
